###  Clearly, we  should not fill up by median. In order to fill up, we should follow the distribution, in this case that is log-normal distribution. To find the parameters, we need to follow the following steps.

(i) Transform the data into log  

(ii) find the mean and standard deviation.






```python
data1["Size_log"]=np.log(data1["Size"]) # transformmed into log and stored as "Size-log"
```

```python
data11=data1.groupby("year")["Size_log"].agg(Mean_Size_log="mean",std_Size_log="std").reset_index() # finding the mean and standard deviation
```

```python
data11 # mean and std, year wise
```

```python
'''
 Here I have randomly genereted the value following the log normal distribution, keeping in mind that no value should
be more than 100 as the maximum value is 100 (original data). If the generated value is more than 100, then it will
be replaced by the median value.
'''

lst=[]
for i in data1["year"]:

    x=np.random.lognormal(data11[data11["year"]==i].Mean_Size_log.tolist()[0],data11[data11["year"]==i].std_Size_log.tolist()[0])
    if x<100:
        y=x
    else:
        y=data1[data1["year"]==i]["Size"].median()
    lst.append(y)
```

```python
data1["Size_new"]=lst # the new column is created as "Size_new" where there is no missing value.
data1["Size"].fillna(data1["Size_new"],inplace=True);
```

```python
# The distribution of "Size" with out missing value
plt.figure(figsize=(8,6))
sns.histplot(data=data1,x="Size",hue="year" ,kde=True)
plt.xlabel("Size")
plt.title("Distribution of 'Size' when missing value filled following the distribution")
plt.show()
```

## Why did you pick the specific chart?
Ans: There are 1461 missing values. In order to fill up these missing values, we need to know the distribution of it.
## What is/are the insight(s) found from the chart?
Clearly, the figure is showing that the distribution is left skewed. By the norms, should have fill up these missing values by median. But one can see the maximum occarance is around 800 but missing value is more than that. Hence We should fill up the missing values following the distribution.

```python
data1["Rating2"]=data1["Rating"]/5 # scaled to get the range [0,1], becuase beta is defined in that range.
```

```python
# get the parameter in beta distribution, need to know the mean and variance
data_rating=data1.groupby("year")["Rating2"].agg(Mean="mean",Std="std").reset_index()
```

```python
data_rating
```

```python
# The function which returns parameters (a,b)
def beta(mu,sigma):
    #input: mu (mean) and sigma (std)
    # output: the parameters (a,b)
    a=((1-mu)/sigma**2-1/mu)*mu**2
    b=a*(1/mu-1)
    return a,b
```

```python
# year-wise, the data is generated
lst2=[]
for year in data1["year"]:
    mu=data_rating[data_rating["year"]==year]["Mean"].tolist()[0]
    sigma=data_rating[data_rating["year"]==year]["Std"].tolist()[0]
    a,b=beta(mu,sigma)

    x=round(5*np.random.beta(a, b),1) #random generated data is multiplied with 5
    #print(x)
    lst2.append(x)
```

```python
# plotting generated data
sns.histplot(data=lst2,kde=True)
plt.title("Generated data of 'Rating'")
plt.show()
```

```python
data1["Rating_New"]=lst2 # in order to fill the missing data, ned to creat a new col.
```

```python
# Lets drop the unnecessary colmuns
data.drop(columns=["Reviews","Installs","Price","Genres",	"Sentiment_Polarity","Sentiment_Subjectivity"],inplace=True);
```

```python
#Here,we will try to find how many positive, neutral and negative sentiment are there.

Apps=list(data["App"].unique())
dict1=dict()
for app in Apps:
    positive=data[(data["App"]==app)&(data["Sentiment"]=="Positive")]["Sentiment"].count()
    negative=data[(data["App"]==app)&(data["Sentiment"]=="Negative")]["Sentiment"].count()
    neutral=data[(data["App"]==app)&(data["Sentiment"]=="Neutral")]["Sentiment"].count()
    lst_senti=[positive,negative,neutral]
    dict1[app]=lst_senti
```

```python
Apps2=list(data["App"])
senti_p=[]
senti_n=[]
senti_neu=[]
for app in Apps2:
    #print(app)
    senti_p.append(dict1[app][0])
    senti_n.append(dict1[app][1])
    senti_neu.append(dict1[app][2])
```

```python
data["Positive_sentiment"]=senti_p # creating "Positive_sentiment" column
data["Negative_sentiment"]=senti_n # creating "Negative_sentiment" column
data["Neutral_sentiment"]=senti_neu ## creating "Neutral_sentiment" column
```

```python
data3=data.drop_duplicates().reset_index() # dropping the duplicates which are generated during the last process.
```
